```python
import json
import os
from collections import defaultdict

# JSON 파일이 있는 기본 폴더 경로
base_folder_path = '/content/drive/MyDrive/KT-project/dataset/label/Train'  # 여기에 기본 폴더 경로를 입력하세요

# 분포를 저장할 딕셔너리 (각 값에 해당하는 폴더 이름들을 저장)
wrinkle_value_folders = defaultdict(list)

# 폴더 내 모든 하위 폴더를 순회
for folder_name in os.listdir(base_folder_path):
    folder_path = os.path.join(base_folder_path, folder_name)

    # 폴더인 경우에만 처리
    if os.path.isdir(folder_path):
        # 해당 폴더의 파일 목록 중 이름이 "1"로 끝나는 JSON 파일 검색
        for file_name in os.listdir(folder_path):
            if file_name.endswith('1.json'):  # 이름이 "1"로 끝나는 파일 필터링
                file_path = os.path.join(folder_path, file_name)

                # JSON 파일을 열어서 "forehead_wrinkle" 값 가져오기
                with open(file_path, 'r', encoding='utf-8') as file:
                    data = json.load(file)
                    forehead_wrinkle = data.get("annotations", {}).get("forehead_wrinkle")

                    # 값이 존재할 경우 딕셔너리에 폴더 이름 추가
                    if forehead_wrinkle is not None:
                        wrinkle_value_folders[forehead_wrinkle].append(folder_name)

                # 첫 번째 JSON 파일만 처리하도록 break
                break

# 분포도 출력
print("Forehead Wrinkle Value Distribution:")
for wrinkle_value, folders in sorted(wrinkle_value_folders.items()):
    # 폴더 이름 정렬
    sorted_folders = sorted(folders)
    count = len(sorted_folders)
    folder_list = ', '.join(sorted_folders)
    print(f'Forehead Wrinkle Value {wrinkle_value} ({count} folders): {folder_list}')
```
